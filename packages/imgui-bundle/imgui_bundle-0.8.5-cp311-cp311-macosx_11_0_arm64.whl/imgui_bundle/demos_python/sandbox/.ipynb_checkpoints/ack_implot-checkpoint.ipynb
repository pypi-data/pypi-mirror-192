```python
import numpy as np
from imgui_bundle import implot, immapp


x = np.arange(0, np.pi * 4, 0.01)
y1 = np.cos(x)
y2 = np.sin(x)


def gui():
    if implot.begin_plot("Plot", (600, 400)):
        implot.plot_line("y2", x, y2)
        implot.end_plot()

immapp.run_nb(gui, thumbnail_height=300)
```

# Getzler's relation (and symmetrizing tautological classes)
In the tautological ring of $\overline{\mathcal{M}}_{1,4}$ there exists a tautological relation between classes of boundary strata of complex codimension $2$, found by Ezra Getzler. See the [original paper](https://www.ams.org/journals/jams/1997-10-04/S0894-0347-97-00246-4/S0894-0347-97-00246-4.pdf) or watch the [ICM talk of Rahul Pandharipande](https://youtu.be/LbNQHE2sJjk?t=2952) for the statement of the relations. Below you find the pictorial representation of the relation, stolen from the slides of the ICM talk:

We input the left-hand side of this relation as a tautological class and verify that it is indeed contained in the span of known tautological relations. 
### Remarks
In the picture above, the markings (in red) are not labeled, indicating that each term stands for the sum of all boundary strata associated to possible labelings. While it's possible to enumerate these by hand, we can use the function ``symmetrize`` that gives the symmetrization of a tautological class.

```python
from admcycles import *

TR = TautologicalRing(1,4)
#TR.sepbdiv(1,[1,2]).rename_legs({1: 3, 2: 1, 3: 2, 4: 4})
Class1 = StableGraph([0,1,0],[[1,2,7],[6,8],[3,4,9]],[(6,7),(8,9)]).to_tautological_class().symmetrize()
Class2 = StableGraph([0,0,1],[[1,2,7],[6,8,3],[4,9]],[(6,7),(8,9)]).to_tautological_class().symmetrize()
Class3 = StableGraph([0,0,1],[[1,2,7],[6,8,3,4],[9]],[(6,7),(8,9)]).to_tautological_class().symmetrize()
Class4 = StableGraph([0,0,1],[[1,2,3,7],[6,8,4],[9]],[(6,7),(8,9)]).to_tautological_class().symmetrize()
Class5 = (1/2)*StableGraph([0,0],[[1,2,3,7],[6,8,4,9]],[(6,7),(8,9)]).to_tautological_class().symmetrize()
Class6 = (1/2)*StableGraph([0,0],[[1,2,3,4,7],[6,8,9]],[(6,7),(8,9)]).to_tautological_class().symmetrize()
Class7 = (1/2)*StableGraph([0,0],[[1,2,7,8],[3,4,6,9]],[(6,7),(8,9)]).to_tautological_class().symmetrize()

(12*3*Class1- 4*12*Class2 -2*6*Class3 +6*4*Class4+ 4*Class5+ Class6-2*3*Class7).is_zero()
```
